```python
def date_to_numeric(date):
    if pd.isna(date):
        return 0
    else:
        try:
            date_obj = pd.to_datetime(date)
            if date_obj.year < 1970 or date_obj.year > 2100:
                return 0
            else:
                return int(date_obj.timestamp())
        except ValueError:
            return 0


df['PrepayP'] = df['PrepayP'].apply(date_to_numeric)
```
