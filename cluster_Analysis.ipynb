```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 5))

# Calculate the density values for each cluster
cluster_densities = []
cluster_labels = []

for i, cluster in enumerate(df['cluster'].unique()):
    density = sns.kdeplot(df.loc[df['cluster'] == cluster, 'BoRace'], ax=ax).get_lines()[0].get_data()
    cluster_densities.append(density)
    cluster_labels.append(cluster)

# Set color palette for clusters
colors = sns.color_palette('Set1', len(cluster_labels))

# Plot the density bar plot with consistent colors for each cluster
for i, cluster_density in enumerate(cluster_densities):
    sns.barplot(x=cluster_density[0], y=cluster_density[1], ax=ax, color=colors[i], alpha=0.7, label=f'Cluster {cluster_labels[i]}')

# Set x-axis labels
ax.set_xticks([])

# Set cluster labels in the top right corner
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, loc='upper right')

# Show the plot
plt.show()

```
